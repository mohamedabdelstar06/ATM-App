```python
from pyspark import SparkContext
```

```python
sc = SparkContext("local", "Task1")
```

```python
rdd1 = sc.textFile("Task1.txt")
```

```python
print("RDD Content:")
print(rdd1.collect())
```

```python
print("\nNumber of elements:", rdd1.count())
```

```python
def split_words(line):
    return line.split()
```

```python
rdd_words = rdd1.map(split_words)
```

```python
print("\nWords in each line:")
print(rdd_words.collect())
```

```python
print("\nNumber of partitions:", rdd1.getNumPartitions())
```
